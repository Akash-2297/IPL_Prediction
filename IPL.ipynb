# current score 

```python
delivary_df['current_score']=delivary_df.groupby('id').cumsum()['total_runs_x']
```

```python
delivary_df
```

# Runs left

```python
delivary_df['runs_left']=delivary_df['total_runs_y']-delivary_df['current_score']
```

```python
delivary_df.head(30)
```

# Wicket_left

```python
delivary_df.columns
```

```python
delivary_df['player_dismissed']=delivary_df['player_dismissed'].fillna("0")
delivary_df['player_dismissed'] = delivary_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
delivary_df['player_dismissed']=delivary_df['player_dismissed'].astype('int')
wicket=delivary_df.groupby('id').cumsum()['player_dismissed'].values
delivary_df['wicket']=10-wicket
```

# BALL LEFT

```python
delivary_df.columns
```

```python
delivary_df['ball_left']=126-(delivary_df['over']*6 + delivary_df['ball'])
```

```python
##curent run rate
delivary_df['crr']=(delivary_df['current_score']*6)/(120-delivary_df['ball_left'])
```

```python
delivary_df['rrr']=(delivary_df['runs_left']*6)/delivary_df['ball_left']
```

```python
delivary_df.head()
```

```python
def result(row):
    return 1 if row.batting_team == row.winner else 0
    
```

```python
delivary_df['result']=delivary_df.apply(result,axis=1)
```

```python
delivary_df.dropna(inplace=True)
```

```python
delivary_df=delivary_df[delivary_df['ball_left']!=0]
```

```python
final_df=delivary_df[['batting_team','bowling_team','ball_left','city','runs_left','wicket','total_runs_y','crr','rrr','result']]
```

```python
##Shuffling the data avoid bias output

final_df=final_df.sample(final_df.shape[0])
```

# MODEL BUILDING

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import  OneHotEncoder
```

```python
x = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
ct=ColumnTransformer([ 
    ('ct',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])],remainder='passthrough')
```

```python
pip=Pipeline(steps=[
    ('step1',ct),('step2',LogisticRegression(solver="liblinear"))
])
```

```python
pip.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred=pip.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```
